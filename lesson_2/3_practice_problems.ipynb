### Problem 3

```python
lst = [10, 9, -6, 11, 7, -16, 50, 8]

sorted_list = sorted(lst, key=str)
print(sorted_list)
```

```python
lst = [10, 9, -6, 11, 7, -16, 50, 8]

sorted_list = sorted(lst, reverse=True, key=str)
print(sorted_list)
```

### Problem 4

```python
def extract_year(book_dict):
    return int(book_dict['published'])

books = [
    {
        'title': 'One Hundred Years of Solitude',
        'author': 'Gabriel Garcia Marquez',
        'published': '1967',
    },
    {
        'title': 'The Book of Kells',
        'author': 'Multiple Authors',
        'published': '800',
    },
    {
        'title': 'War and Peace',
        'author': 'Leo Tolstoy',
        'published': '1869',
    },
]

sorted_list = sorted(books, key=extract_year)
print(sorted_list)
```
