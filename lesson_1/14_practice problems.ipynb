### Problem 1

```python
fruits = ("apple", "banana", "cherry", "date", "banana")

# one way
print(fruits.count("banana"))

# another way
count = 0
for item in fruits:
    if item == "banana":
        count += 1
print(count)
```

### Problem 4

```python
names = ["Fred", "Barney", "Wilma", "Betty", "Pebbles", "Bambam"]
name_positions = {}
for index, name in enumerate(names):
    name_positions[name] = index
print(name_positions)

# {"Fred": 0, "Barney": 1, "Wilma": 2, "Betty": 3, "Pebbles": 4, "Bambam": 5}
```

### Problem 5

```python
ages = {
    "Herman": 32,
    "Lily": 30,
    "Grandpa": 5843,
    "Eddie": 10,
    "Marilyn": 22,
    "Spot": 237,
}

# one way
list_of_ages = list(ages.values())
print(sum(list_of_ages))

# another way
print(sum(ages.values()))
```

### Problem 7

```python
words = ['ant', 'bear', 'cat']
selected_words = []
for word in words:
    if len(word) > 3:
        selected_words.append(word)

print(selected_words)

# ['bear']
```

### Problem 8

```python
statement = "The Flintstones Rock"
end_dict = {}

for char in statement:
    end_dict[char] = statement.count(char)
del end_dict[' ']

print(end_dict)
```

### Problem 10

```python
dictionary = {'a': 'ant', 'b': 'bear'}
print(dictionary.popitem())

# ('b', 'bear')  -- the popped item is returned by .popitem; it's the last
# item because since Python 3.7, LIFO order is preserved with dictionaries
```
